```python
#import libraries
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
#past lottery data ミニロト(MINI LOTO) of Japan
URL = "http://sougaku.com/miniloto/data/list1/"
page = requests.get(URL)

'''
pass page.content instead of page.text to avoid problems with character encoding. 
The .content attribute holds raw bytes, which can be decoded better than the text representation
'''
soup = BeautifulSoup(page.content, "html.parser")
```

```python
'''
inspect website 
<table class="bun_box2" ...>
.
.
.
<tr>
    <td class="w6"> 第1000回 </td>
    <td class="w5"> 3 </td> 本数字
    <td class="w5"> 8 </td> 本数字
    <td class="w5"> 14 </td> 本数字
    <td class="w5"> 18 </td> 本数字
    <td class="w5"> 26 </td> 本数字
    <td class="w4"> 7 </td> B数字
    <td> B </td> セット
</tr>
'''
```

```python
table = soup.find('table', attrs={"class":"bun_box2"})
table_rows = table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)


df = pd.DataFrame(res, columns = ["past_drawings", "number_1", "number_2", "number_3","number_4","number_5", "bonus_number", "set", "main_number odd:even", "total_main", "all_number odd:even", "total_all", "Details"])

#inspect df
df.describe()
```

```python
#drop non-necessary columns
to_drop = ['main_number odd:even',
                'total_main', 
                'all_number odd:even', 
                'total_all', 
                'Details']

df.drop(to_drop, inplace=True, axis=1)

df.head()
```
